## Data preparation


Let's start by importing the <em>pandas</em> and the Numpy libraries.


```python
import pandas as pd
import numpy as np
```

### We start by reading and understanding data...


```python
concrete_data = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
print("Data Header")
display(concrete_data.head())
print("\n","*"*70,"\n")
print("Data Summary")
display(concrete_data.describe())
print("\n","*"*70,"\n")
print("Missing values")
concrete_data.isnull().sum()
```

```python
concrete_data.shape
```

So, there are approximately 1000 samples to train our model on. Because of the few samples, we have to be careful not to overfit the training data.


The data looks very clean and is ready to be used to build our model.


### We split data into predictors and target


The target variable in this problem is the concrete sample strength. Therefore, our predictors will be all the other columns.


```python
concrete_data_columns = concrete_data.columns

predictors = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']] # all columns except Strength
predictors.head()
```

<a id="item2"></a>


```python
target = concrete_data['Strength'] # Strength column
target.head()
```

Let's save the number of predictors to *n_cols* since we will need this number when building our network.


```python
n_cols = predictors.shape[1] # number of predictors
```

### Let's implement the baseline model according to the assignment

- One hidden layer of 10 nodes, and a ReLU activation function

- Use the adam optimizer and the mean squared error  as the loss function.

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
# define regression model
def regression_model():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

# build the model
model = regression_model()
```

#### 1. Randomly split the data into a training and test sets by holding 30% with sklearn

<a id='item34'></a>


```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size=0.3, random_state=42)
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

#### 2. Train the model on the training data using 50 epochs.

```python
model.fit(X_train, y_train, epochs=50, verbose=1)
```

#### 3. Evaluate the model on the test data and compute the mean squared error

##### 3.1. Evaluation with keras

```python
print("loss = ", model.evaluate(X_test, y_test))
```

##### 3.2. Evaluation with sklearn

```python
# predictions
y_pred = model.predict(X_test)

from sklearn.metrics import mean_squared_error

print("loss = ", mean_squared_error(y_test, y_pred))
```

#### 4. Repeat steps 1 - 3, 50 times, i.e., create a list of 50 mean squared errors.


```python
mse_list=[]
for i in range(50):
    X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size=0.3, random_state=i)
    model.fit(X_train, y_train, epochs=50, verbose=1)
    loss = model.evaluate(X_test, y_test)
    mse_list.append(loss)
   
```

```python
print("We reapeat the 3 first steps and create a list including", len(mse_list), "MSE")
```

#### 5.Report the mean and the standard deviation of the mean squared errors.

##### 5.1. Mean of the 50 mean squared errors

```python
print("Mean of the mean squared errors = ", np.mean(mse_list))
```

##### 5.2. Standard deviation of the 50 mean squared errors

```python
print("The standard deviation of the mean squared errors = ", np.std(mse_list))
```
